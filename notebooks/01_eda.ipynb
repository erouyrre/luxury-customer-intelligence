```python
import pandas as pd
from pathlib import Path

CURRENT_DIR = Path().resolve()
RAW_FILE = CURRENT_DIR.parent.joinpath("data", "raw", "online_retail.xlsx")
df = pd.read_excel(RAW_FILE)
print(df.shape)
print(df.head())
print(df.info())
print(df.isnull().sum())
```

```python
# Regarde les valeurs uniques d'InvoiceNo
print(df["InvoiceNo"].sample(10))

# Regarde les valeurs uniques de StockCode
print(df["StockCode"].sample(10))

# Regarde les valeurs commençant par C de InvoiceNo
df[df["InvoiceNo"].astype(str).str.startswith("C")].head(10)
```

```python
# Regarde les valeurs négatives de UnitPrice
print(df[df["UnitPrice"] <= 0].shape)

print(df["UnitPrice"].describe())
```

```python
# Regarde les bordures de la colonne InvoiceDate
print(df["InvoiceDate"].min())
print(df["InvoiceDate"].max())

# Regarde la répartition des Country
print(df["Country"].value_counts().head(10))
```
